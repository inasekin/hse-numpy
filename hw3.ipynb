1. Как зовут владельца, у недвижимости которого за отчетный период была наибольшая разница между максимальной и минимальной ценой?

Цены в календаре записаны как строки типа '$100.00'. Нужно преобразовать их в числа.
Сначала пробовал просто убрать символы и применить astype(float), но получил NaN везде. Оказалось проблема в пустых значениях.

```python
calendar['price_num'] = pd.to_numeric(
    calendar['price'].str.replace('$', '', regex=False).str.replace(',', '', regex=False),
    errors='coerce'
)

print(f"\nNaN: {calendar['price_num'].isna().sum()}")
print(f"Ненулевых: {calendar['price_num'].notna().sum()}")
```

теперь ищем min и max и далее разницу

```python
price_stats = calendar.groupby('listing_id')['price_num'].agg(['min', 'max'])
price_stats['diff'] = price_stats['max'] - price_stats['min']

max_diff_id = price_stats['diff'].idxmax()
max_diff_value = price_stats['diff'].max()

print(f"Listing ID: {max_diff_id}")
print(f"Разница: ${max_diff_value:.2f}")

owner = listings[listings['id'] == max_diff_id]['host_name'].values[0]
print(f"Владелец: {owner}")
```

```python
calendar['max_price'] = calendar.groupby('listing_id')['price_num'].transform('max')
calendar['min_price'] = calendar.groupby('listing_id')['price_num'].transform('min')
calendar['diff'] = calendar['max_price'] - calendar['min_price']

max_diff_id2 = calendar.loc[calendar['diff'].idxmax(), 'listing_id']
owner = listings[listings['id'] == max_diff_id2]['host_name'].values[0]
print(f"Владелец: {owner}")
```

```python
print("agg")
%timeit calendar.groupby('listing_id')['price_num'].agg(['min', 'max'])

print("transform")
%timeit calendar.groupby('listing_id')['price_num'].transform('max') - calendar.groupby('listing_id')['price_num'].transform('min')
```

3. Коррелируют ли средняя цена на недвижимость за всё время наблюдений и количество отзывов на неё? Иными словами, если средняя цена квартиры больше, то на неё оставляют больше или меньше отзывов?

```python
avg_price = calendar.groupby('listing_id')['price_num'].mean().reset_index()
avg_price.columns = ['id', 'avg_price']

data = listings[['id', 'number_of_reviews']].merge(avg_price, on='id')
data = data.dropna()

data.head()
```

```python
len(data)
```

```python
corr_pearson = data['avg_price'].corr(data['number_of_reviews'])
print(f"Корреляция Пирсона {corr_pearson:.4f}")
```

корреляция отрицательная, значит чем дороже объявление, тем меньше отзывов.


```python
avg_price_series = calendar.groupby('listing_id')['price_num'].mean()
data2 = listings.set_index('id')[['number_of_reviews']].join(avg_price_series.rename('avg_price'))
data2 = data2.dropna()

corr_pearson2 = data2['avg_price'].corr(data2['number_of_reviews'])
print(f"{corr_pearson2:.4f}")
```

Что все её гости отмечали в качестве плюсов сдаваемой недвижимости?

```python
top_host_id = namesake_count.index[0][0]
comments = namesakes[namesakes['host_id'] == top_host_id]['comments']

comments.head(3)
```

```python
import re
import string

PUNCT_RE = re.compile(f'[{re.escape(string.punctuation)}]')

def preprocess_text(text):
    text = PUNCT_RE.sub('', text)
    return text.split()

all_text = ' '.join(comments.astype(str)).lower()
word_list = preprocess_text(all_text)

word_counts = pd.Series(word_list).value_counts()

stop_words = {
    'the', 'was', 'and', 'a', 'to', 'in', 'very', 'great', 'of', 
    'is', 'but', 'few', 'as', 'we', 'i', 'he', 'she', 'it', 'they', 
    'are', 'were', 'be', 'been', 'am', 'or', 'on', 'at', 'with', 
    'for', 'about', 'from', 'by', 'that', 'this', 'those', 'these',
    'its', 'an', 'not', 'days', 'host', 'boston'
}

filtered_word_counts = word_counts[~word_counts.index.isin(stop_words)]
filtered_word_counts.head(20)
```
