9. Создайте столбец с датой контакта. Дата должна соответствовать информации из столбцов month и day_of_week и относиться к 2021 году (в этом задании может быть, разумеется, много верных ответов, выведите любой)

```python
df['contact_date'] = pd.to_datetime(
    '2021-' + df['month'] + '-01',
    format='%Y-%b-%d'
)
```

```python
day_map = {'mon': 0, 'tue': 1, 'wed': 2, 'thu': 3, 'fri': 4}
first = pd.to_datetime('2021-' + df['month'], format='%Y-%b')
df['contact_date'] = first + pd.to_timedelta(
    (df['day_of_week'].map(day_map) - first.dt.dayofweek) % 7, unit='D'
)
```
